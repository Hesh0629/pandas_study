# 판다스 내장 그래프 도구

plot의 경우 행 인덱스를 X축으로 봄

```python
path='/content/drive/MyDrive/pandas_example_data/남북한발전전력량.xlsx'
df=pd.read_excel(path)
df_ns=df.iloc[[0,5],3:]
df_ns.index=['South','North']
#df_ns.columns=df_ns.columns.map(int)
print(df_ns)
df_ns.plot()
```

따라서 x축에 연도를 두고싶으면 전치행렬 사용

```python
tdf_ns=df_ns.T
print(tdf_ns.head(10))
tdf_ns.plot()
```

값의 빈도를 보여주는 히스토그램

```python
tdf_ns.plot(kind='hist') # X축은 발전량을 나타내고, Y축은 해당 발전량 구간안에 들어오는 연도의 빈도
```

두 데이터 간의 관계를 보여주는 산점도

```python
path='/content/drive/MyDrive/pandas_example_data/auto-mpg.csv'
df=pd.read_csv(path,header=None)
df.columns=['mpg','cylinders','displacement','horsepower','weight','acceleration','model year','origin','name']
df.plot(x='weight',y='mpg',kind='scatter') # X축을 weight, Y축을 연비(mpg)라 할 때 상관성을 볼 수 있음 (이경우에는 음의 상관관계 -0.831741)
asdf=df.loc[:,['mpg','weight']]
asdf.corr()
```
